```python

class Node:
    def __init__(self, value):
        self.value = value


class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)
        print(f"Pushed → {item}")

    def pop(self):
        if not self.is_empty():
            item = self.items.pop()
            print(f"Popped → {item}")
            return item
        return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        print("Current Stack:", self.items)



def dfs_recursive(graph, node_value, visited, stack):
    if node_value not in visited:
        stack.push(node_value)
        visited.append(node_value)
        print(f"Visited → {visited}")


        neighbors = graph.get(node_value, [])

      
        visit_neighbors(graph, neighbors, visited, stack)

       
        stack.pop()
        stack.display()


def visit_neighbors(graph, neighbors, visited, stack):
    if not neighbors:   # Base condition
        return
    first = neighbors[0]
    dfs_recursive(graph, first, visited, stack)       
    visit_neighbors(graph, neighbors[1:], visited, stack)  


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

start_node = Node('A')
stack = Stack()
visited = []

dfs_recursive(graph, start_node.value, visited, stack)

print("Visited Order:", visited)
```
